```python
import pandas as pd
import seaborn as sns
```

```python
vendas = pd.read_excel("varejo.xlsx")
vendas.head()
```

```python
vendas.info()
```

```python
vendas.groupby("idcanalvenda").idcompra.nunique().reset_index()
```

```python
vendas["idcanalvenda"] = vendas["idcanalvenda"].str.replace("APP", "Aplicativo")
vendas.groupby("idcanalvenda").idcompra.nunique().reset_index()
```

```python
vendas.groupby("estado").idcompra.nunique().reset_index()
```

```python
vendas["Nome_Departamento"] = vendas["Nome_Departamento"].str.replace(" ", "_")
vendas.groupby("Nome_Departamento").idcompra.nunique().reset_index()
```

```python
vendas.info()
```

```python
vendas[vendas["estado"].isnull()]
```

```python
nulos = vendas.dropna()
nulos.info()
```

```python
vendas["estado"].fillna("MS", inplace=True)
```

```python
vendas.info()
```

```python
media_preço = vendas["Preço"].mean()
```

```python
media_preço
```

```python
vendas["Preço"].fillna(media_preço, inplace=True)
```

```python
vendas.info()
```

```python
vendas[["Preço","Preço_com_frete"]].describe()
```

```python
preço_errado = vendas.query("Preço > Preço_com_frete")
preço_certo = vendas.query("Preço < Preço_com_frete")
```

```python
preço_certo.describe()
```

```python
preço_certo.head()
```

```python
preço_certo.query("Nome_Departamento == 'Esporte_e_Lazer' and estado == 'SP'").describe()
```

```python
cliente = pd.read_excel("cliente_varejo.xlsx")
```

```python
cliente.head()
```

```python
preço_certo.head()
```

```python
 cliente[["idade", "renda"]].describe()
```

```python
cliente.info()
```

```python
cliente = cliente.astype({"renda":"float"})
```

```python
vendas_cliente = preço_certo.merge(cliente, how= "left", on="cliente_Log")
vendas_cliente.head()
```

```python
agg_idcanal_renda = round(vendas_cliente.groupby("idcanalvenda")["renda"].agg("mean").sort_values(ascending=False).reset_index(), 2)
agg_idcanal_renda
```

```python
agg_idade_bandeira = round(vendas_cliente.groupby("bandeira")["idade"].agg("mean").sort_values(ascending=False).reset_index(), 2)
agg_idade_bandeira
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.Figure(figsize=(7,4))

plt.bar(agg_idade_bandeira["bandeira"],agg_idade_bandeira["idade"], color= "green")

plt.ylabel("Média de idade")

plt.title("Idade Média por Bandeira", loc="center", fontsize=18, fontweight="bold", color= "black");
```

```python
fig = plt.Figure(figsize=(7,4))

plt.bar(agg_idcanal_renda["idcanalvenda"],agg_idcanal_renda["renda"], color= "Blue")

plt.ylabel("Média Canal de venda")

plt.title("Renda Média por Canal de Venda", loc="center", fontsize=18, fontweight="bold", color= "black");
```

```python
preço_certo.groupby("Data").idcompra.nunique().sort_values(ascending=False).reset_index()
```

```python
venda_por_data = preço_certo.groupby("Data").idcompra.nunique().sort_values(ascending=False).reset_index()
venda_por_data
```

```python
# Assuming venda_por_data is a pandas DataFrame
fig = plt.figure(figsize=(22, 14))

plt.xlabel("Data da Venda")
plt.ylabel("Quantidade de vendas")
x = venda_por_data["Data"]
y = venda_por_data["idcompra"]

plt.plot(x, y, color="green", linewidth=2)
plt.show()
```

```python
venda_dept_preco= round(vendas_cliente.groupby("Nome_Departamento")["Preço_com_frete"].agg("mean").sort_values(ascending=False).reset_index(), 2)
venda_dept_preco
```

```python
plt.figure(figsize=(32, 12))

plt.bar(venda_dept_preco["Nome_Departamento"], venda_dept_preco["Preço_com_frete"], color="blue")

plt.ylabel("Preço Médio")
plt.title("Preço Médio por Departamento", loc="center", fontsize=18, fontweight="bold", color="black")

plt.show()
```
